### État de l'art : 

- Référence : Compton, Todd (2017). *Who Wrote the Beatle Songs?  A History of Lennon-McCartney* :

Comme le remarque Todd Compton, la plupart des chansons des Beatles ont été signées par les deux principaux chanteurs du groupe : John Lennon et Paul McCartney. Ce n'est pas un mystère cependant que ces chansons, pour une partie, n'ont été écrites que par un seul des deux. </br>
La question de l'attribution des textes a été rendue encore plus compliquée du fait des Beatles eux-mêmes, qui ont cultivé la mésinformation par rapport à ces questions. De même, ce point aveugle tient sûrement aussi du fait que l'apport majeur des Beatles et la rupture qu'ils constituent dans la culture populaire ne tient pas forcément à leurs textes. </br>
Certaines chansons relèvent d'une véritable collaboration, écrites pendant des sessions de trois heures, "*eyeball to eyeball*" dit John Lennon. </br>
D'autres ont été écrites séparément par John Lennon ou Paul McCartney et l'un aidait l'autre à terminer d'écrire la chanson. Selon T. Compton, il s'agit de la forme la plus commune de leur collaboration. </br>
Dans d'autres cas, des textes étaient écrits séparément par les deux chanteurs, puis rassemblés ensuite pour construire une chanson unique (ex : *Baby You're a Rich Man*"). </br>

Lorsque l'on se penche sur les interviews donnés par les deux chanteurs, on se rend compte que ceux-ci se sont souvent contredits sur ces questions, l'un vis-à-vis de l'autre, mais aussi au sein de leur propre discours personnel. Pour ne prendre qu'un exemple parmi les nombreux développés par T. Compton, la chanson *Eleanor Rigby*, attribuée en grande partie à McCartney mais dont Lennon affirme dans un interview de 1970 avoir écrit près de 70% de la chanson, propos contredits par un des amis de McCartney, Pete Shotton, présent lors de la session, qui affirme que Lennon n'a pratiquement rien fait sur cette chanson. Lennon lui-même, dans un interview de 1981, explique qu'il a aidé à écrire une demie ligne de *Eleanor Rigby*, se contredisant avec ses propos tenus dix ans plus tôt. De même, dans un autre interview, répondant à la question de savoir quelles chansons étaient les plus marquantes dans sa collaboration avec McCartney au niveau de l'écriture, Lennon cite *Eleanor Rigby*. Pour une demie ligne ? C'est peu probable...

Une vision diachronique sur les différentes périodes des Beatles peut s'avérer nécessaire pour tâcher de mieux comprendre cette collaboration.

- Première période : "__the Beatles era__" : Brian Epstein, l'agent des Beatles, leur conseilla rapidement d'affirmer que toutes leurs chansons résultaient d'une écriture collaborative. En 1964, McCartney affirme : "*Sometimes maybe he'll write a whole song himself, or I will, but we always say that we've both written it*" ; et en 1966, d'affirmer : "*We both work on all our songs*". Au fil du temps cependant, et après la mort d'Epstein en 1967, cette position de revendication d'une écriture résolument collaborative se dissipa progressivement, à mesure qu'ils arrivaient aux albums de maturité, comme *Revolver and Sgt. Pepper*.
- Deuxième période : "__after the breakup__" : très peu de temps après la rupture en 1970-1971, Lennon donna des interviews qui remirent en question l'attribution des chansons attribuées au couple Lennon-McCartney. Cependant, ces propos semblent davantage relever du ressentiment de Lennon vis-à-vis de McCartney à cette époque et de leurs désaccords. Compton écrit : "*During this period, John and Paul were publicly and heatedly disagreeing about many things, and attacked each other in interviews, at the same time that they were discussing autorship of songs. Naturally, bias would be present during a time of such heated emotional conflict*" (p. 6).
- Troisième période : "__1980's to the present__" : cette troisième période se caractérise par une forme de retour sur les propos tenus pendant la deuxième période. Dans sa biographie *Paul McCartney : Many Years From Now*, McCartney affirme n'être en désaccord avec Lennon sur l'attribution de seulement deux chansons. En vérité, ces désaccords sont bien plus nombreux, mais l'on remarque que ces interviews ont eu tendance à reprendre la première posture de l'écriture purement collaborative.

On note donc que l'histoire de l'écriture collaborative entre les deux auteurs se doublent des histoires et destins personnelles de ceux-ci, de  problématiques tenant à leur relation amicale, des tensions qui ont pu naître entre eux, mais aussi de l'image commerciale qui a été entretenue par leur agent, préférant mettre en avant la création collective plutôt que l'écriture individuelle au sein du groupe. </br>
Si Todd Compton mène une étude en confrontant les sources, les différents interviews et propos des deux chanteurs, il convient pour nous de nous demander si une analyse à partir des textes et du style de ces chansons à l'attribution douteuse ne nous permettrait pas, pour certaines chansons tout du moins, de démêler le style propre de chacun des deux chanteurs.

```python
corpus_clean = corpus

# Retrait des NaN dans les paroles :

corpus_clean.dropna(subset = ["Lyrics"], inplace=True)
corpus_clean[0:1]
```

**Plusieurs remarques à faire ici** :

    - On remarque évidemment l'absence d'homogénéisation des différentes observations 
    (==> nettoyage nécessaire des données).
    - Également les retours chariot qui peuvent poser pb.

**À modifier** :

    - McCartney\n (with Lennon)
    - Lennon\n McCartney
    - Lennon\n (with McCartney)
    - McCartney (with Lennon)
    - Lennon\n and McCartney

==> 'Lennon-McCartney'

```python
## Regex de nettoyage :

## Sélection de la colonne, retrait retour chariot et homogénéisation des obs

corpus_clean['Songwriter(s)'] = corpus_clean['Songwriter(s)'].replace('\n','', 
                                                                      regex=True) # retour chariot

corpus_clean['Songwriter(s)'] = corpus_clean['Songwriter(s)'].replace('McCartney (with Lennon)',
                                                                      'Lennon-McCartney', regex=False) # Regex = False parenthèse

corpus_clean['Songwriter(s)'] = corpus_clean['Songwriter(s)'].replace('Lennon McCartney',
                                                                      'Lennon-McCartney', regex=False)

corpus_clean['Songwriter(s)'] = corpus_clean['Songwriter(s)'].replace('Lennon (with McCartney)',
                                                                      'Lennon-McCartney', regex=False)

corpus_clean['Songwriter(s)'] = corpus_clean['Songwriter(s)'].replace('McCartney (with Lennon)',
                                                                      'Lennon-McCartney', regex=False)

corpus_clean['Songwriter(s)'] = corpus_clean['Songwriter(s)'].replace('Lennon and McCartney',
                                                                      'Lennon-McCartney', regex=False)

# Aperçu du résultat :

corpus_clean[0:15]
```

```python
# Construction :
from spacy.lang.en import English
nlp = English()
# Create a Tokenizer with the default settings for English
# including punctuation rules and exceptions
tokenizer = nlp.tokenizer
```

```python
import spacy

# Petit réglage pour permettre d'écrire sur les données...
# (sécurité panda)

pd.options.mode.chained_assignment = None  # default='warn' (Cf. https://stackoverflow.com/questions/20625582/how-to-deal-with-settingwithcopywarning-in-pandas)

corpus_test = corpus_clean_select

corpus_test['words'] = corpus_test['Lyrics'].apply(lambda x: nlp.tokenizer(str(x)))
# df['sents'] = df['text'].apply(lambda x: list(nlp(x).sents))
# Réf : https://stackoverflow.com/questions/46981137/tokenizing-using-pandas-and-spacy

# Note that nlp by default runs the entire SpaCy pipeline, which includes part-of-speech tagging, parsing and named entity recognition. You can significantly speed up your code by using nlp.tokenizer(x) instead of nlp(x), or by disabling parts of the pipeline when you load the model. E.g. nlp = spacy.load('en', parser=False, entity=False).

# Transform data using spaCy
# nlp = spacy.load("en_core_web_sm")
# WARNING: takes a long time!
# corpus_test['words'] = corpus_test['Lyrics'].apply(lambda x: [sent.text for sent in nlp(x).sents])

corpus_test.head()
```

```python
# https://towardsdatascience.com/tokenize-text-columns-into-sentences-in-pandas-2c08bc1ca790
corpus_test = corpus_test.explode("words", ignore_index=True)
corpus_test.head()
# The explode() function is used to transform each element of a list-like to a row, 
# replicating the index values. Returns: Series- Exploded lists to rows; index will 
# be duplicated for these rows.
```

```python
# À faire : téléchargement du modèle :   

# python -m spacy download en_core_web_sm

nlp = spacy.load("en_core_web_sm")

## Méthode :

## Cf. https://stackoverflow.com/questions/44395656/applying-spacy-parser-to-pandas-dataframe-w-multiprocessing

## Spacy is highly optimised and does the multiprocessing for you. 
## As a result, I think your best bet is to take the data out of 
## the Dataframe and pass it to the Spacy pipeline as a list rather 
## than trying to use .apply directly.
## You then need to the collate the results of the parse, and put 
## this back into the Dataframe. 

lemma = []
pos = []

for doc in nlp.pipe(corpus_test['words'].astype('unicode').values, batch_size=50):
    if doc.has_annotation("DEP"):
        #tokens.append([n.text for n in doc])
        lemma.append([n.lemma_ for n in doc])
        pos.append([n.pos_ for n in doc])
    else:
        # We want to make sure that the lists of parsed results have the
        # same number of entries of the original Dataframe, so add some blanks in case the parse fails
        # tokens.append(None)
        lemma.append(None)
        pos.append(None)
        
# corpus_test['tokens'] = tokens
corpus_test['lemma'] = lemma
corpus_test['pos'] = pos

```
